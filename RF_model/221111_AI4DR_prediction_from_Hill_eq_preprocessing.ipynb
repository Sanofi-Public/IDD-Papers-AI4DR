```python
import math
import shutil
import pandas as pd
import numpy as np
from numpy import loadtxt
import glob
import sys
```

```python
from operator import add
```

```python
from hillfit import fitting
```

```python
test_file = '221108_AI4DR_Test.csv'
out_test_file = '221112_AI4DR_Test_Hill_parameters.csv'
```

```python
test_df = pd.read_csv(test_file, sep=';')
```

```python
test_df.info()
```

```python
test_df['xArray_l'] = test_df['xArray'].map(lambda x: list(eval(x)))
```

```python
test_df.head(2)
```

```python
test_df['Y1_l'] = test_df['Y1_str'].map(lambda x: list(eval(x)))
test_df['Y2_l'] = test_df['Y2_str'].map(lambda x: list(eval(x)))
```

```python
mean_Y = []
for y1,y2 in zip(test_df['Y1_l'],test_df['Y2_l']):
    y_tmp = list( map(add, y1, y2) )
    y_tmp = [ x/2 for x in y_tmp]
    mean_Y.append(str(y_tmp))
```

```python
test_df['Y_mean_str'] = mean_Y
```

test_df.head(2)

```python
curr_dir = '/root/workdir/221111/AI4DR'
r_2_list = []
top_list = []
bottom_list = []
ec50_list = []
nh_list = []
for xx,yy in zip(test_df['xArray'],test_df['Y_mean_str']):
    xx = list(eval(xx))
    yy = eval(yy)
    yy.reverse()
    hf = fitting.HillFit(xx, yy)
    try:
        hf.fitting(log_x = True, generate_figure = False)
        r_2_list.append(hf.r_2)
        top_list.append(hf.top)
        bottom_list.append(hf.bottom)
        ec50_list.append(hf.ec50)
        nh_list.append(hf.nH)
    except:
        r_2_list.append(np.nan)
        top_list.append(np.nan)
        bottom_list.append(np.nan)
        ec50_list.append(np.nan)
        nh_list.append(np.nan)
    dirs = glob.glob(f'{curr_dir}/Hillfit-reg*/')
    for d in dirs:
        shutil.rmtree(d)
        #print(d)
    #break
```

```python
test_df['r2'] = r_2_list
test_df['top'] = top_list
test_df['bottom'] = bottom_list
test_df['ec50'] = ec50_list
test_df['nh'] = nh_list
```

```python
test_df = test_df[['row ID', 'CurveType', 'Classe', 'Classe_Index', 'CATOP', 'CANB', 'CASIG', 'CANT', 'CAHS', 'CNA', 'P', 'NT', 'LS', 'BA', 'BTOX', 'W', 'LU', 'r2', 'top', 'bottom', 'ec50', 'nh']]
```
